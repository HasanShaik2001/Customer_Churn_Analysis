```python

plt.figure(figsize = (3,4))

grouped = df.groupby(['SeniorCitizen', 'Churn']).size().reset_index(name='Count')

# Step 1: Calculate % of total per SeniorCitizen group
total_per_group = grouped.groupby('SeniorCitizen')['Count'].transform('sum')
grouped['Percent'] = (grouped['Count'] / total_per_group) * 100

# Step 2: Pivot for stacking
pivot = grouped.pivot(index='SeniorCitizen', columns='Churn', values='Percent').fillna(0)

# Step 3: Plot using stacked bars
colors = sns.color_palette("Set2", n_colors=2)
labels = pivot.columns.tolist()
x = pivot.index.tolist()
bottom_vals = [0] * len(pivot)

fig, ax = plt.subplots(figsize=(8,5))

for i, label in enumerate(labels):
    heights = pivot[label].values
    ax.bar(x, heights, bottom=bottom_vals, label=label, color=colors[i])
    # Add % text
    for j, (b, h) in enumerate(zip(bottom_vals, heights)):
        if h > 0:
            ax.text(j, b + h/2, f'{h:.1f}%', ha='center', va='center', fontsize=10)
    bottom_vals = [b + h for b, h in zip(bottom_vals, heights)]

# Step 4: Customizations
ax.set_xticks([0, 1])
ax.set_xticklabels(['No', 'Yes'])
ax.set_title('Customer Churn % by Senior Citizen')
ax.set_ylabel('% of Total')
ax.set_xlabel('Senior Citizen')
ax.legend(title='Churn')
plt.tight_layout()
plt.show()
```

# comparitively a greater percentage of senior citizens have been churned out

```python
columns = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 
           'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup',
           'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
           'Contract', 'PaperlessBilling', 'PaymentMethod', 'Churn']

# Plot
plt.figure(figsize=(18, 30))  # Adjust size as needed

for i, col in enumerate(columns, 1):
    plt.subplot(6, 3, i)  # 6 rows, 3 columns of subplots
    sns.countplot(x=col, data=df, palette='Set2',hue="Churn")
    plt.title(col)
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

# the absence of services like Online Security, Tech Support, and Device Protection is associated with higher churn, highlighting the potential of bundling such features to improve retention. Payment method plays a role too—those using electronic checks show a noticeably higher churn rate than users paying via credit cards or bank transfers. While streaming services (TV or movies) don't have as strong an impact, customers who subscribe to them still show a slight tendency to churn, potentially reflecting optional services that are easier to drop.
